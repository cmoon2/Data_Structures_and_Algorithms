# Find a Missing Element

## Problem

Given a first array of non-negative integers. A second array is generated by shuffling the elements of the first array and deleting a random element. Find which element is missing in the second array. 

For example,

Input:
    
    find_missing_element([1,2,3,4,5,6,7],[3,7,2,1,4,6])

Output:

    5 



```python
def find_missing_element(arr1, arr2):
    
    # The first array should have at least one element
    if len(arr1) < 1:
        return
    
    for num in arr2:
        arr1.remove(num)
        
    return arr1[0]
```

```python
arr1 = [1,2,3,4,5,6,7]
arr2 = [3,7,2,1,4,6]
find_missing_element(arr1, arr2)
```

## Complexity

O(N^2)

```python
def find_missing_element(arr1, arr2):
    
    # The first array should have at least one element
    if len(arr1) < 1:
        return
    
    # Sort the input arrays
    arr1.sort()
    arr2.sort()
    
    # Make list of pairs of elements that are in the same indices in the sorted arrays
    # Find a pair that has different values
    for num1, num2 in zip(arr1, arr2):
        if num1 != num2:
            return num1
    
    # Otherwise return the last element
    return arr1[-1]
```

```python
arr1 = [1,2,3,4,5,6,7]
arr2 = [3,7,2,1,4,6]
find_missing_element(arr1, arr2)
```

O(N log N)
